```
fig = plt.figure(figsize= (12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('alaska.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('ALGIERIA.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('antarktyda.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('bilblioteka.jpg'))
```

```
fig = plt.figure(figsize= (12, 10))

for idx, img in enumerate(['alaska.jpg', 'ALGIERIA.jpg', 'antarktyda.jpg', 'bilblioteka.jpg']):
  plt.subplot(221 + idx)
  plt.imshow(mpimg.imread(img))
```

```
fig = plt.figure(figsize= (12, 10))

for idx, img in enumerate(['alaska.jpg', 'ALGIERIA.jpg', 'antarktyda.jpg', 'bilblioteka.jpg']):
  plt.subplot(141 + idx)
  plt.imshow(mpimg.imread(img))
  plt.axis('off')
```
