```
# rzadziej używana biblioteka, są lepsze, ta nie jest interaktywna, ale czasem się przydaje
import matplotlib
matplotlib.__version__
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
import seaborn as sns  #trochę ładniejsze wykresy
sns.set()
```

```
fig = plt.figure()
fig.show()
```

Wykres liniowy

```
_ = plt.plot([0, 1], [0, 3])
```

```
_ = plt.plot([0, 1], [0, 3], label = 'line', color = 'green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```
_ = plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

TWORZENIE WIELU WYKRESÓW

```
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
```

```
plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'ro')
```

```
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'ro')

fig.show()  #w Colabie jest to domyslnie wyświetlane więc nie musimy tego pisać
```

WYKRESY SŁUPKOWE

```
plt.bar(x=[0, 1, 2, 3], height= [1.2, 3.2, 1.2, 5.2])
```

```
plt.bar(x=['a', 'b', 'c', 'd'], height= [1.2, 3.2, 1.2, 5.2])
```

```
plt.barh(y=['a', 'b', 'c', 'd'], width= [1.2, 3.2, 1.2, 5.2])
```

WYKRESY PUNKTOWE

```
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
```

```
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
