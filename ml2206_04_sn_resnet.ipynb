```
# inicjalizacja modelu / SN
model = resnet50.ResNet50()
```

```
img = image.load_img("doda.jpg", target_size=(224,224))
img
```

```
img = image.load_img("putin.jpg", target_size=(224,224))
img
```

```
# zamaiana na postac macierzy binarnej/ bitmapy
x = image.img_to_array(img)
```

```
x.shape  #trzy wymiary\; szerokość, wysokość i kolor
```

```
# powiększamy ilośc wymiarow o +1 bo musimy stworzyć kolekcję obrazów - czyli brakuje nr obrazu w kolekcji (tu tylko 1 element w kolekcji)
x = np.expand_dims(x, axis=0)    #rozciągamy wegług pierwszej osi - wymiar oznaczający listę obrazow
x.shape
```

```
x = resnet50.preprocess_input(x)
```

```
predict = model.predict(x)
```

```
resnet50.decode_predictions(predict, top=5)
```
